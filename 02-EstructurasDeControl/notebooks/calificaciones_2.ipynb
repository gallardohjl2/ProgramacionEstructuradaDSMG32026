# Programa de Calificaciones

Este programa calcula el promedio de tres calificaciones y evalua si un alumno **Aprobo** o **No Aprobo**

## Definicion de Variables

```python
calif1 = 0.0
calif2 = 0.0
calif3 = 0.0
promedio = 0.0
```

## Entrada de Datos

```python
calif1 = float(input("Introduce la Calificación 1"))
calif2 = float(input("Introduce la Calificación 2"))
calif3 = float(input("Introduce la Calificación 3"))
```

## Proceso: Calculo de Promedio

```python
promedio = (calif1 + calif2 + calif3) / 3
```

## Comparación con una Estructura selectiva doble

```python
if promedio >= 7.0:
    print("Aprobado")
else:
    print("No Aprobado")
```
